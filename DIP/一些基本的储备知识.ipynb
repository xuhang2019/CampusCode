## cv2的数学计算 包括numpy
#### cv2.64F
64位浮点类型，经常用于uint8负数的时候转换,经典用法：

```
sobelx8u = cv.Sobel(img,cv.CV_8U,1,0,ksize=5)
# Output dtype = cv.CV_64F. Then take its absolute and convert to cv.CV_8U
sobelx64f = cv.Sobel(img,cv.CV_64F,1,0,ksize=5)
abs_sobel64f = np.absolute(sobelx64f)
sobel_8u = np.uint8(abs_sobel64f)
```

## np的经典用法
### np用=无法修改内存，必须使用.copy()才能真正的复制一份，要不然全部相当于在复制地址，包括他的slice特性

#### np.where
- 代替二维数组的遍历
- np.where(data<0,data+180,data)
    - 用于选择部分选项，小于0的部分+180
- X, Y = np.where((M > np.max(M) * 0.3)&(alpha>45)&(alpha<135))
    - X，Y返回两种数组下标，再通过g[X,Y] = 1可以直接赋值。
    
#### np.clip
- 限制数组中的超过的部分的值，并处理成一定的值。比如限制于大于1的部分为1。、
