```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
source_path = 'source/log_小车惯性数据.txt'
```

```python
with open(source_path, "r", encoding="utf8") as f:
    # x.strip() for x in lines
    lines = []
    for x in f:
        if(x[0] == 'R' or x[0] == 'A'):
            lines.append(x)
    
    lines2 = [x for x in f if(x[0] == 'R' or x[0] == 'A')] # 不行，我也不知道为什么

aacx = []

for line in lines:
    line = line.replace(": ",r",")
    linesp = line.split(',')
    line_A = []
    cnt = 0
    if(linesp == None):
        continue
    for i,cell in enumerate(linesp):
        if(len(cell) < 1):
            continue
        if cell[0] == 'A':
            line_A.append(i)
    for i,ind in enumerate(line_A):
        if(len(line_A) == 2):
            aacx.append(int(linesp[line_A[0]+5]))
            break
        elif(len(line_A) >= 2):
            if i < len(line_A)-1:
                if((line_A[i+1]) - line_A[i] > 5):
                    aacx.append(int(linesp[line_A[i]+5]))
```

```python
plt.figure()
plt.plot(range(0,len(aacx)),aacx)
```
